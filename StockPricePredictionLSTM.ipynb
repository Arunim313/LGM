```python
data["Date"] = data.index
data = data[["Date" , "Open" , "High" , "Low" , "Close" , "Total Trade Quantity" , "Turnover (Lacs)"]]
data.reset_index(drop = True , inplace=True)
data.tail()
```

```python
# GRAPH REPRESENTATION
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=data["Date"] , open=data["Open"] , high = data["High"] , low = data["Low"] , close = data["Close"] )])
figure.update_layout(title = "TATA GLOBAL Stock Price Analysis" , xaxis_rangeslider_visible = False)
figure.show()
```

```python
correlation = data.corr()
print(correlation["Close"].sort_values(ascending = False))
```

```python
# TRAINING LSTM FOR PREDICTION
x = data[["Open","High", "Low" , "Turnover (Lacs)"]]
y = data["Close"]
x = x.to_numpy()
y = y.to_numpy()
y = y.reshape(-1 , 1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# PREPARING NEURAL NETWORK FOR LSTM
from keras.models import Sequential
from keras.layers import Dense, LSTM
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (xtrain.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(xtrain, ytrain, batch_size=1, epochs=30)
```

```python
import numpy as np
#features = [Open, High, Low, Total trade quantity , Turnover (Lacs)]
features = np.array([[177.089996, 180.419998, 177.070007, 74919600]])
model.predict(features)
```
